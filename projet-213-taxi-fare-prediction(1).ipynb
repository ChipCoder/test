```python
#read train set
train = pd.read_csv("../input/train.csv", nrows = 5_000_000)
print("shape of train data", train.shape)
train.head()
```

```python
# Basic Stats of the data set
train.describe()
```

From the above description, we can found that:
1. min fare_amount < 0
2. max passenger_count = 208
So, next we do the dataset cleaning...

```python
#fare amount >= 0
print("old size: %d" %len(train))
train = train[train.fare_amount >= 0]
print("new size: %d" %len(train))
```

```python
#Then, check the missing data
train.isnull().sum()
print("old size: %d" %len(train))
train = train.dropna(how = "any", axis = 0)
print("new size: %d" %len(train))
```

```python
#read test dataset
test = pd.read_csv("../input/test.csv")
print("shape of train data", test.shape)
test.describe()
```

```python
test.isnull().sum()
print("old size: %d" %len(test))
test = test.dropna(how="any", axis = 0)
print("new size: %d" %len(test))
```

```python
min_lgt = min(test.pickup_longitude.min(), test.dropoff_longitude.min())
min_lat = min(test.pickup_latitude.min(), test.dropoff_latitude.min())
max_lgt = max(test.pickup_longitude.max(), test.dropoff_longitude.max())
max_lat = max(test.pickup_latitude.max(), test.dropoff_latitude.max())
POS = (min_lgt, min_lat, max_lgt, max_lat)
print(POS)
```

```python
def filter_within_test_boundary(dataset, POS):
    return (dataset.pickup_longitude >= POS[0]) & (dataset.pickup_longitude <= POS[2]) &\
    (dataset.dropoff_longitude >= POS[0]) & (dataset.dropoff_longitude <= POS[2]) &\
    (dataset.pickup_latitude >= POS[1]) & (dataset.pickup_latitude <= POS[3]) &\
    (dataset.dropoff_latitude >= POS[1]) & (dataset.dropoff_latitude <= POS[3])
```

```python
print("old size: %d" %len(train))
train = train[filter_within_test_boundary(train, POS)]
print("new size: %d" %len(train))
```

```python
plt.figure(figsize=(10,6))
train[train.passenger_count > 7].passenger_count.hist(bins=20)
plt.figure(figsize=(10,6))
train[train.passenger_count > 6].passenger_count.hist(bins=20)
```

```python
print("old size: %d" %len(train))
train = train[train.passenger_count <= 6]
print("new size: %d" %len(train))
```
